## 01
- Dado un diccionario con una lista de nombres de películas y el año concatenado, crear una función que devuelva un DataFrame con dos columnas:
  - Columna 1, el nombre de la película sin el año
  - Columna 2, el año de la película

```python
# Diccionario de entrada: cada película tiene el formato "nombre - año"
movies = {'Name': ["Matrix - 1999", "Cadena Perpetua - 1994", "Parasitos - 2019", "Vengadores, Endgame - 2019", "Interstellar - 2014",
                   "Django Desencadenado - 2012", "El Padrino - 1972", "El Caballero Oscuro - 2008", "Erase Una Vez En America - 1984",
                   "El Senyor de los Anillos, La Comunidad del Anillo - 2001"]}
# DataFrame de referencia con el resultado correcto (nombre y año separados)
expected_output = pd.DataFrame({'Name':["Matrix", "Cadena Perpetua", "Parasitos", "Vengadores, Endgame", "Interstellar", "Django Desencadenado",
                                        "El Padrino", "El Caballero Oscuro", "Erase Una Vez En America", "El Senyor de los Anillos, La Comunidad del Anillo"],
                                'Year':["1999", "1994", "2019", "2019", "2014", "2012", "1972", "2008", "1984", "2001"]})
# Llamamos a nuestra función para transformar el diccionario en DataFrame
output = extract_year(movies)
# Mostramos el DataFrame resultante en pantalla
display(output)
# Verificamos que el resultado sea idéntico al esperado; si no, lanza AssertionError
assert(expected_output.equals(output))
```

```python
# Your solution HERE !!!!
# Función principal: recibe un diccionario y devuelve un DataFrame con Name y Year
def extract_year(movies):
    # Función auxiliar: extrae solo el año del título completo
    def split_year(title):
        index = title.index('-')  # Busca la posición del guion separador
        return title[index:].replace('-','').strip()  # Toma lo que va después del guion y limpia espacios
    
    # Función auxiliar: extrae solo el nombre de la película (sin el año)
    def remove_year(title):
        index = title.index('-')  # Posición del guion
        return title[:index-1].strip()  # Toma el texto anterior al guion (y al espacio) y limpia espacios

    df_movies = pd.DataFrame(movies)  # Convierte el diccionario en DataFrame
    df_movies['Year'] = df_movies['Name'].apply(split_year)  # Crea columna Year aplicando split_year a cada título
    df_movies['Name'] = df_movies['Name'].apply(remove_year)  # Sobrescribe Name dejando solo el nombre de la película
    
    return df_movies  # Devuelve el DataFrame final con dos columnas
```

```python
# Solución alternativa del ejercicio 01
def extract_year(dicci):
  dicci2={}  # Diccionario vacío para guardar las dos listas resultantes
  peli=[]  # Lista donde iremos guardando los nombres de película
  ano=[]  # Lista donde iremos guardando los años
  for i in dicci['Name']:  # Recorremos cada título del diccionario de entrada
    titulo,year=i.split(" - ")  # Separamos nombre y año usando " - " como separador
    peli.append(titulo)  # Añadimos el nombre a la lista de películas
    ano.append(int(year))  # Añadimos el año convertido a entero

  dicci2={'Name':peli,'Year':ano}  # Construimos el nuevo diccionario con ambas columnas
  frame=pd.DataFrame(dicci2)  # Creamos el DataFrame a partir del diccionario
  return frame  # Devolvemos el DataFrame final
```

```python
# Otra solución alternativa del ejercicio 01
def extract_year(movies):
  nombres = []  # Lista para almacenar los nombres sin año
  anios = []  # Lista para almacenar los años
  for titulo in movies['Name']:  # Iteramos cada cadena "película - año"
    partes = titulo.rsplit(' - ', 1)  # Divide por el último " - " (útil si el título tiene comas o guiones)
    nombre = partes[0]  # Primera parte: nombre de la película
    anio = partes[1]  # Segunda parte: año
    nombres.append(nombre)  # Guardamos el nombre
    anios.append(anio)  # Guardamos el año

  return pd.DataFrame({'Name': nombres, 'Year': anios})  # Creamos y devolvemos el DataFrame
```

```python
# Solución compacta del ejercicio 01 usando funciones lambda
def extract_year(movies):
    df_movies = pd.DataFrame(movies)  # Convertimos el diccionario en DataFrame
    # Extraemos el año: partimos por '-' y nos quedamos con la segunda parte, sin espacios
    df_movies['Year'] = df_movies['Name'].apply(lambda x : x.split('-')[1].strip())
    # Extraemos el nombre: partimos por '-' y nos quedamos con la primera parte, sin espacios
    df_movies['Name'] = df_movies['Name'].apply(lambda x : x.split('-')[0].strip())
    return df_movies  # Devolvemos el DataFrame con columnas Name y Year
```

## 02
- Crear una función que reciba un array de NumPy y devuelva otro en el cual estén marcados como True los elementos duplicados (a partir de la segunda ocurrencia), y como False los no repetidos o las primeras ocurrencias de los duplicados.

```python
# Array de prueba con números, algunos repetidos
numbers = np.array([4, 4, 4, 3, 8, 1, 9, 6, 1, 5, 9, 4, 0, 2])
# Resultado esperado: True solo desde la 2ª aparición de cada valor duplicado
expected_output = np.array([False, True, True, False, False, False, False, False, True, False, True, True, False, False])
# Llamamos a nuestra función que marca duplicados
output = find_duplicates(numbers)
# Imprimimos el array booleano resultante
print(output)
# Comprobamos que coincida exactamente con lo esperado
assert(np.array_equal(expected_output, output))
```

```python
# Your solution HERE !!!!
# Función que marca True los duplicados a partir de su segunda aparición
def find_duplicates(numbers):
    serie = pd.Series(numbers)  # Convertimos el array NumPy en una Serie de Pandas
    return serie.duplicated()  # duplicated() devuelve True para elementos repetidos (excepto la 1ª vez)
```

```python
# Solución alternativa del ejercicio 02 con NumPy
def find_duplicates(numbers):
    _, idx = np.unique(numbers, return_index = True)  # idx = posiciones de la primera aparición de cada valor único
    result = np.zeros_like(numbers, dtype = bool)  # Creamos un array de False del mismo tamaño que numbers
    result[idx] = True  # Marcamos True en las primeras apariciones
    result = np.invert(result)  # Invertimos: True queda solo en duplicados (2ª aparición en adelante)
    return result  # Devolvemos el array booleano final
```

```python
# Solución alternativa del ejercicio 02 con set y bucle
def find_duplicates(numbers):
  visto = set()  # Conjunto para recordar valores ya encontrados
  output = []  # Lista donde guardaremos True/False para cada posición
  for value in numbers:  # Recorremos cada número del array
    if value in visto:  # Si ya lo vimos antes, es un duplicado
      output.append(True)  # Marcamos True (2ª aparición o posterior)
    else:
      visto.add(value)  # Si es la primera vez, lo guardamos en el set
      output.append(False)  # La primera aparición no se marca como duplicado
  return np.array(output)  # Convertimos la lista a array NumPy y la devolvemos
```

```python
# Otra solución alternativa del ejercicio 02 con diccionario
def find_duplicates(numeros):
  dicci={}  # Diccionario para registrar valores ya vistos
  resul=np.full(numeros.shape, False, dtype=bool)  # Array de False con la misma forma que el input
  for idx, valor in enumerate(numeros):  # Recorremos índice y valor de cada elemento
    if valor in dicci:  # Si el valor ya apareció antes
      resul[idx]=True  # Marcamos esa posición como duplicado
    else:
      dicci[valor]=True  # Registramos la primera aparición
  return resul  # Devolvemos el array booleano
```

## 03
- Resolución de dos sistemas de ecuaciones
    - $ A : \begin{cases} -9x +  4y = 20 \\ -7y + 16x = 80                         \end{cases}$
    
    - $ B : \begin{cases} x - 2y + 3z = 7 \\ 2x + y + z = 4 \\ -3x + 2y -2z = -10 \end{cases}$
- Verifica por código los resultados

```python
# Your solution HERE !!!!
import math  # Para comparar números decimales con tolerancia (isclose)
import numpy.linalg as linalg  # Módulo de álgebra lineal de NumPy

# Sistema A: matriz de coeficientes de las ecuaciones -9x+4y=20 y 16x-7y=80
a = np.array([ [-9,4], [16,-7] ])
# Vector de términos independientes del sistema A
b = np.array([20,80])

print(a)  # Mostramos la matriz A
print(b)  # Mostramos el vector b
x,y = linalg.solve(a,b)  # Resolvemos el sistema Ax=b y obtenemos x e y
print(f'x={x} y={y}')  # Imprimimos la solución

# Verificamos que la solución cumple la primera ecuación: -9x + 4y = 20
print(math.isclose(-9*x + 4*y, 20))
# Verificamos que cumple la segunda ecuación: 16x - 7y = 80
print(math.isclose(16*x - 7*y, 80))

# Sistema B: matriz 3x3 para x - 2y + 3z = 7, 2x + y + z = 4, -3x + 2y - 2z = -10
A2 = np.array([[1, -2, 3], [2, 1, 1], [-3, 2, -2]])
# Vector de términos independientes del sistema B
b2 = np.array([7, 4, -10])

x,y,z = np.linalg.solve(A2, b2)  # Resolvemos el sistema 3x3
print(f'x={x} y={y} z={z}')  # Mostramos las tres incógnitas

# Comprobamos ecuación 1: x - 2y + 3z = 7
print(math.isclose(+1*x-2*y+3*z,7))
# Comprobamos ecuación 2: 2x + y + z = 4
print(math.isclose(2*x+1*y+z,4))
# Comprobamos ecuación 3: -3x + 2y - 2z = -10
print(math.isclose(-3*x+2*y-2*z,-10))
```

```python
# Your solution HERE !!!!
# Creamos un DataFrame 10x10 con enteros aleatorios entre 0 y 20 (21 excluido)
df = pd.DataFrame(np.random.randint(21, size = (10,10)))
display(df)  # Mostramos el DataFrame original
# Reemplazamos valores menores que 5 por NaN; inplace=True modifica df directamente
df.where(df >= 5, np.nan, inplace=True)
display(df)  # Mostramos el DataFrame después del reemplazo
# Contamos cuántos NaN hay en cada fila (axis=1 recorre columnas por fila)
print(df.isnull().sum(axis = 1))
```

## 07 - 1pt
- Escribir un generador `frange` que simule el comportamiento de la función `arange` de NumPy
- Esta función cada vez que se invoque debe devolver valores espaciados uniformemente a partir de un valor inicial. Se debe proveer a dicha función un valor inicial para la secuencia de números y un valor para el step
- Debe soportar valores númericos en coma flotante
- La función potencialmente se puede llamar un número infinito de veces
- Usando plain Python, sin ninguna librería

```python
import math  # Necesario para comparar floats con tolerancia

init = 10.6  # Valor inicial de la secuencia
step = 0.8  # Salto entre números consecutivos
g = frange(init, step)  # Creamos el generador infinito

iters = 10  # Número de valores que queremos obtener
numbers = [next(g) for n in range(iters)]  # Sacamos 10 valores del generador
for i, n in enumerate(numbers):  # Comprobamos cada valor generado
    assert(math.isclose(n, init + (step * i)))  # Debe cumplir: valor = inicio + step * índice
print(numbers)  # Mostramos la secuencia generada
```

```python
# Your solution HERE !!!!
# Generador infinito similar a arange: devuelve init, init+step, init+2*step, ...
def frange(init, step):
    while True:  # Bucle infinito: el generador puede llamarse muchas veces
        yield init  # Devuelve el valor actual y pausa la ejecución
        init += step  # Suma el paso para preparar el siguiente número
```

## 08
- Crear una función que reciba una Serie de Pandas y devuelva esa misma serie, reemplazando los espacios en blanco por el carácter menos frecuente en dicha serie.

```python
# Serie de prueba con espacios que deben reemplazarse
serie = pd.Series(list('bba cabc faabba aacbbfe'))
# Resultado esperado: espacios sustituidos por 'e' (el carácter menos frecuente)
expected_output = pd.Series(list('bbaecabcefaabbaeaacbbfe'))
# Llamamos a la función de limpieza
output = clean_whitespace_replace(serie)
# Verificamos que el resultado sea exactamente el esperado
assert(expected_output.equals(output))
```

```python
# Your solution HERE !!!!
# Reemplaza espacios por el carácter menos frecuente de la Serie
def clean_whitespace_replace(serie):
    less_freq = serie.value_counts().idxmin()  # Cuenta frecuencias y toma el carácter con menor recuento
    serie[serie == ' '] = less_freq  # Sustituye cada espacio por ese carácter
    return serie  # Devuelve la Serie modificada
```

```python
from collections import Counter  # Counter cuenta cuántas veces aparece cada elemento

# Solución alternativa del ejercicio 08
def clean_whitespace_replace (seri):
  texto=''.join(seri.dropna().astype(str))  # Unimos todos los caracteres de la Serie en un solo string
  frecu=Counter(i for i in texto if i != '')  # Contamos la frecuencia de cada carácter no vacío
  mefrecu=min(frecu,key= frecu.get)  # Obtenemos el carácter con menor frecuencia
  seridif=seri.astype(str).apply(lambda s: s.replace(' ' ,mefrecu))  # Reemplazamos espacios en cada elemento
  return seridif  # Devolvemos la Serie transformada
```
