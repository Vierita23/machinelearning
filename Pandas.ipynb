# 2 Series

• Datos unidimensionales (similar a NumPy)
• Elementos + índices modificables

```python
# Creamos una Serie: vector 1D con índice automático (0, 1, 2...)
countries = pd.Series(['Spain','Andorra','Gibraltar','Portugal','France'])
# Mostramos la Serie completa con índice y valores
print(countries)
```

```python
# Creamos Serie con índice personalizado: las etiquetas no tienen que ser 0,1,2...
countries = pd.Series(['Spain','Andorra','Gibraltar','Portugal','France'],
                      # index=range(10,60,10) asigna etiquetas 10, 20, 30, 40, 50
                      index=range(10,60,10))
# Verificamos que cada país queda asociado a su etiqueta de índice
print(countries)
```

```python
# Los índices pueden ser strings u otros tipos, no solo enteros
football_cities = pd.Series(['Barcelona','Madrid','Valencia','Sevilla'],
                            # Índices semánticos 'a','b','c','d' para acceso por etiqueta
                            index=['a','b','c','d'])
# Mostramos la Serie con índice de letras
print(football_cities)
```

```python
# Atributos de la Serie: nombre de la serie y nombre del índice
football_cities.name = 'Ciudades con dos equipos en primera' # nombrar la Serie
# index.name describe la columna de etiquetas (como cabecera del índice)
football_cities.index.name = 'Id' # Describir los índices
# Mostramos la Serie con metadatos de nombre aplicados
print(football_cities)
```

```python
# Acceso por posición entera: iloc[2] devuelve el 3er elemento (índice 0-based)
print(football_cities.iloc[2])
# Acceso por etiqueta semántica: ['c'] busca el valor cuyo índice es 'c'
print(football_cities['c'])
# Comprobamos que ambos accesos devuelven el mismo valor (Valencia)
print(football_cities['c'] == football_cities.iloc[2])
```

# 3 Tratamiento similar a ndarray

```python
# Selección múltiple por etiquetas: devuelve sub-Serie con índices 'a' y 'c'
print(football_cities[ ['a','c'] ])
# Equivalente por posición: iloc[[0,3]] toma el 1er y 4to elemento
print(football_cities.iloc[[0, 3]])
```

```python
# Convertimos la sub-Serie a lista de Python (solo los valores)
lista = list(football_cities[:'c'])
# Mostramos la lista resultante ['Barcelona', 'Madrid', 'Valencia']
print(lista)
# Confirmamos que el tipo es list, no Series
print(type(lista))
```

```python
# Convertimos la sub-Serie a ndarray de NumPy para operaciones vectoriales
import numpy as np
# np.array extrae los valores y los convierte en array NumPy
cities = np.array(football_cities[:'c'])
# Mostramos el array NumPy resultante
print(cities)
# Verificamos el tipo: ndarray permite operaciones matemáticas eficientes
print(type(cities))
```

```python
# Convertimos la sub-Serie a diccionario {índice: valor}
lista = dict(football_cities[:'c'])
# Muestra pares clave-valor, ej: {'a': 'Barcelona', 'b': 'Madrid', ...}
print(lista)
```

```python
# Máscaras booleanas: filtrar elementos que cumplan una condición
fibonacci = pd.Series([0, 1, 1, 2, 3, 5, 8, 13, 21])
# Mostramos la Serie completa de Fibonacci
print(fibonacci)
# Creamos máscara: True donde el valor es mayor que 10
mask = fibonacci > 10
# La máscara es una Serie de booleanos alineada con el índice original
print(mask)
# Aplicamos la máscara: solo devuelve valores donde mask es True (13 y 21)
print(fibonacci[mask])
# Serie de comparación con los dos valores filtrados
dst = pd.Series([13,21])
# Mostramos la Serie de referencia para comparar
print(dst)
# equals() compara valores e índice: False porque dst tiene índice 0,1 y fibonacci 7,8
dst.equals(fibonacci)
# Extraemos los valores >10 y los guardamos en fb
fb = fibonacci[mask]
# reset_index(drop=True) renumera el índice a 0,1,... sin conservar el original
fb.reset_index(drop=True, inplace=True)
# Ahora fb tiene índice [0,1] con valores [13, 21]
print(fb)
# Tras reindexar, fb y dst tienen mismos valores e índice → True
dst.equals(fb)
```

```python
# Las Series de Pandas son compatibles con funciones universales de NumPy
import numpy as np
# np.sum suma todos los elementos de la Serie (ignora NaN por defecto)
print(np.sum(fibonacci))
```

```python
# np.where reemplaza valores según condición: útil para imputar NaN
distances = pd.Series([12.1,np.nan,12.8,76.9,6.1,7.2])
# Si no es NaN → conserva el valor; si es NaN → sustituye por 2
valid_distances = np.where(pd.notnull(distances),distances,2)
# Resultado: array NumPy con NaN reemplazados por 2
print(valid_distances)
# np.where devuelve ndarray, no Series
print(type(valid_distances))
```

• Interpretar el índice como clave

```python
# Consulta condicional: verificar si una etiqueta existe antes de acceder
# print(serie['Marcos'])
if 'Marcos' in serie:
    # Solo se ejecuta si 'Marcos' sigue en el índice (aquí ya fue eliminado)
    print(serie['Marcos'])
# Mostramos el estado actual de la Serie tras las operaciones anteriores
print(serie)
```

## 3.2 Operaciones entre series

```python
# Suma de Series: se alinean por índice, no por posición
# suma de valores con el mismo índice (NaN si no aparece en ambas)
serie1 = pd.Series([10,20,30,40], index=range(4) )
# serie2 solo tiene índices 0,1,2 → el índice 3 de serie1 quedará NaN
serie2 = pd.Series([1,2,3], index=range(3) )
# Suma elemento a elemento donde coinciden índices; NaN donde falta pareja
suma = serie1 + serie2
# Resultado: [11, 22, 33, NaN]
print(suma)
```

```python
# resta de series (similar a la suma)
print(serie1 - serie2)
```

```python
# Pre-filtrado: operar primero y luego reemplazar NaN resultantes
result = serie1 + serie2
# Detectamos valores nulos (NaN)
# isnull() crea máscara True donde hay NaN; asignamos 0 en esas posiciones
result[pd.isnull(result)] = 0 # mask con isnull()
# Ahora el índice 3 tiene 0 en lugar de NaN
print(result)
```

### 3.2.1 Diferencias entre Pandas Series y diccionario

• Diccionario, es una estructura que relaciona las claves y los valores de forma arbitraria.
• Series, estructura de forma estricta listas de valores con listas de índice asignado en la posición.
• Series, es más eficiente para ciertas operaciones que los dicionarios.
• En las Series los valores de entrada pueden ser listas o Numpy arrays.
• En Series los índices semánticos pueden ser integers o caracteres, en los valores igual.
• Series se podría entender entre una lista y un diccionario Python, pero es de una dimensión.

## 4.1 Modificar DataFrames

```python
# Añadir columna nueva: asignar un valor escalar rellena todas las filas
diccionario = { "Nombre" : ["Marisa","Laura","Manuel"],
                # Columna Edad con las edades de las tres personas
                "Edad" : [34,29,12] }
# Creamos DataFrame con columnas extra vacías (NaN) para Nacionalidad, Profesion, Direccion
frame = pd.DataFrame(diccionario, columns=['Nacionalidad', 'Nombre', 'Edad','Profesion', 'Direccion'])
# Asignamos 'Desconocida' a toda la columna Direccion (broadcasting de escalar)
frame['Direccion'] = 'Desconocida'
# Tabla con la nueva columna rellena uniformemente
display(frame)
```

```python
# Lista con una dirección distinta para cada fila del DataFrame
lista_direcciones = ['Rue 13 del Percebe, 13', 'Evergreen Terrace, 3', 'Av de los Rombos, 12']
```

```python
# Asignamos la lista: cada elemento se alinea con una fila por posición
frame['Direccion'] = lista_direcciones
# Cada fila recibe su dirección correspondiente de la lista
display(frame)
```

```python
# Añadir fila con loc: requiere valores para todas las columnas existentes
# añadir fila (requiere todos los valores)
user_2 = ['Alemania','Klaus',20, 'none','Desconocida']
# loc[3] crea/asigna la fila con índice 3 usando la lista en orden de columnas
frame.loc[3] = user_2
# DataFrame con la nueva fila de Klaus en la posición 3
display(frame)
```

```python
# Eliminar filas y columnas: drop no modifica in-place por defecto
# eliminar fila (similar a Series)
frame = pd.DataFrame(diccionario,columns=['Nacionalidad', 'Nombre', 'Edad', 'Profesion'])
# drop(2) elimina la fila con índice 2; hay que reasignar porque devuelve copia
frame = frame.drop(2) # por qué necesitamos reasignar el frame?
# Tabla sin la fila de Manuel (índice 2)
display(frame)
# drop con axis=1 elimina columnas; inplace=True modifica el DataFrame original
frame.drop('Nombre', axis = 1, inplace = True)
# DataFrame sin la columna Nombre
display(frame)
```

```python
# del también elimina columnas, como alternativa a drop(axis=1)
#eliminar columna
del frame['Profesion']
# DataFrame final sin la columna Profesion
display(frame)
```

```python
# acceder a la traspuesta (como una matriz)
display(frame.T)
```

```python
# Doble bucle: primero filas (values), luego cada celda de la fila
# iterar sobre filas y luego sobre cada valor?
for values in frame.values:
    # Bucle interno: recorre cada valor individual de la fila actual
    for value in values:
        # Imprime cada celda una a una (incluyendo NaN)
        print(value)
```

## 4.3 Indexación y slicing con DataFrames

```python
# Lista de dicts con claves distintas: Pandas rellena NaN donde falta clave
d1 = {'ciudad':'Valencia', 'temperatura':10, 'o2':1}
# d2 no tiene 'o2' → esa celda será NaN en el DataFrame
d2 = {'ciudad':'Barcelona', 'temperatura':8}
# d3 solo tiene ciudad y temperatura
d3 = {'ciudad':'Valencia', 'temperatura':9}
# d4 añade humedad pero no o2 ni co2
d4 = {'ciudad':'Madrid', 'temperatura':10, 'humedad':80}
# d5 tiene humedad y co2
d5 = {'ciudad':'Sevilla', 'temperatura':15, 'humedad':50, 'co2':6}
# d6 tiene todas las claves posibles del conjunto
d6 = {'ciudad':'Valencia', 'temperatura':10, 'humedad':90, 'co2':10}
# Agrupamos los 6 registros en una lista para crear el DataFrame
ls_data = [d1, d2, d3, d4, d5, d6] # lista de diccionarios
# index='abcdef' etiqueta cada fila; columnas = unión de todas las claves
df_data = pd.DataFrame(ls_data, index = list('abcdef'))
# Tabla heterogénea: NaN donde cada dict no tenía esa clave
display(df_data)
```

```python
# iloc: acceso por posición entera [fila, columna]
# Acceso a un valor concreto por indice posicional [row, col]
print(df_data.iloc[1,1])
# iloc[:3,:4] → primeras 3 filas y 4 primeras columnas
display(df_data.iloc[:3,:4])
# loc con escalares: fila 'a', columna 'temperatura' → valor único
display(df_data.loc['a', 'temperatura'])
# loc con rangos semánticos: filas hasta 'c', columnas hasta 'o2' (inclusivos)
display(df_data.loc[:'c', :'o2'])
# loc con slice de columnas: de 'temperatura' a 'o2' en filas hasta 'c'
display(df_data.loc[:'c', 'temperatura':'o2'])
# loc con lista de columnas: todas las filas, solo 'ciudad' y 'o2'
display(df_data.loc[:, ['ciudad','o2']])
```

```python
# Acceso por nombre de columna: df['col'] devuelve una Series
# indexación con nombre de columna (por columnas)
print(df_data['ciudad']) # --> Series
# Doble corchete [['ciudad','o2']] devuelve DataFrame con esas columnas
display(df_data[['ciudad', 'o2']])
```

```python
# df[0] busca una COLUMNA llamada 0, no la fila 0 → KeyError
# indexación con índice posicional (no permitido!). Esto busca columna.
try:
    # Intentamos acceder con entero como si fuera posición de fila
    df_data[0]
# Capturamos KeyError: no existe columna con nombre 0
except KeyError as e:
    # Mostramos el tipo de error y el mensaje para entender el fallo
    print(type(e).__name__ + ':', e)
```

```python
# indexar por posición con 'iloc'
print(df_data.iloc[0]) # --> Series de la primera fila (qué marca los índices)
```

```python
# indexar semántico con 'loc'
df_data.loc['a'] # --> Series de la fila con índice 'a'
```

```python
# indexar semántico con 'loc'
df_data.loc[:'b'] # --> DataFrame de la fila con índice 'a' y 'b'
```

```python
# Selección por etiqueta de índice (semántica)
df_data.loc[:'b'].loc[:,["o2", "humedad"]] # slicing anidado
```

```python
# SettingWithCopyWarning: modificar una vista afecta al DataFrame original
# si se modifica una porcion del dataframe se modifica el dataframe original (referencia)
display(df_data)
# loc['a'] devuelve una vista (o copia) de la fila 'a' como Series
serie = df_data.loc['a']
# Mostramos la fila extraída
print(serie)
# Modificar serie.iloc[2] cambia también df_data (problema de referencia)
serie.iloc[2] = 3000 # setting with copy warning!!!
# El valor 3000 aparece en df_data original
display(df_data)
# .copy() crea una copia independiente para modificar sin afectar el original
df_2 = df_data.loc['a'].copy()
# Ahora la modificación solo afecta a df_2
df_2.iloc[2] = 3000
# df_2 modificado con o2=3000
display(df_2)
# df_data conserva su valor original en esa celda
display(df_data)
```

```python
# ambos aceptan 'axis' como argumento
# df_data.iloc(axis=1)[0] # --> todos los valores asignados a la primera columna 'ciudad'
df_data.loc(axis=1)['ciudad'] # --> equivalente frame['ciudad']
```

```python
# Índice no ordenado: fila con index=0 no es necesariamente la primera fila
# qué problema puede tener este fragmento?
frame = pd.DataFrame({"Name" : ['Carlos','Pedro'], "Age" : [34,22]}, index=[1,0])
# Carlos tiene index=1 (primera fila visual), Pedro tiene index=0
display(frame)
```

```python
# iloc[0] = primera fila por posición; loc[0] = fila con etiqueta 0
# por defecto, pandas interpreta índice posicional --> error en frames
# cuando hay posible ambigüedad, utilizar loc y iloc
print('Primera fila
')
# iloc[0] devuelve Pedro (posición 0 en el DataFrame visual)
print(frame.iloc[0])
# Imprimimos separador para distinguir ambos resultados
print('
Elemento con index 0
')
# loc[0] devuelve la fila cuya etiqueta de índice es 0 (Pedro también aquí)
print(frame.loc[0])
```

## 4.4 Objeto Index de Pandas

```python
# pd.Index: objeto inmutable que representa el índice de filas/columnas
# Contrucción de índices
ind = pd.Index([2, 3, 5, 23, 26])
# ind[3] accede al 4to elemento del índice → valor 23
print(ind[3])
# Slicing del índice: cada 2 posiciones → [2, 5, 26]
print(ind[::2])
```

```python
# Podemos pasar un pd.Index al crear el DataFrame como etiquetas de fila
# usar un objeto index al crear dataframe
frame = pd.DataFrame({"Name" : ['Carlos','Pedro', 'Manolo', 'Luis', 'Alberto'], "Age" : [34,22,15,55,23]}, index=ind)
# Filas etiquetadas con 2, 3, 5, 23, 26 en lugar de 0-4
display(frame)
```

```python
# Los objetos Index son inmutables: no se puede cambiar un elemento
# Son inmutables! No se modifican los datos.
try:
    # Intentar asignar ind[3]=8 lanza TypeError
    ind[3] = 8
# Capturamos TypeError: los índices no admiten asignación
except TypeError as e:
    # Mostramos el error para confirmar la inmutabilidad
    print(type(e).__name__ + ':', e)
```

```python
# set_index convierte una columna en el nuevo índice del DataFrame
# change index column
frame = pd.DataFrame({"Name" : ['Carlos','Pedro', 'Manolo', 'Luis', 'Alberto'], "Age" : [34,22,15,55,23]}, index=ind)
# DataFrame con índice numérico personalizado antes del cambio
display(frame)
# Age pasa a ser el índice; la columna Age desaparece como columna
frame.set_index('Age', inplace=True)
# Ahora las filas se identifican por edad; Name es la única columna
display(frame)
```

## 4.5 Slicing

```python
# Solución correcta: copiar el slice antes de modificar y reasignar
# filtrar edad incorrecta (mínimo 18) Fixing it by Copying the slice
users_dataset = pd.read_csv(ruta_users, sep='::', index_col=0,
                           # Recargamos con nombres de columnas explícitos
                           header=None, names=['UserID','Gender','Age','Occupation','Zip-code'],
                           # engine='python' para el separador '::'
                           engine='python')
# Subconjunto de menores de edad (vista del original)
under_age = users_dataset[users_dataset['Age'] == 1]
# .copy() crea copia independiente para modificar sin warning
under_age_copy = under_age.copy()
# Estado antes de anular la edad
display(under_age_copy.head())
# Anulamos Age con NaN en la copia (seguro, no afecta users_dataset aún)
under_age_copy['Age'] = np.nan
# Copia modificada con Age=NaN
display(under_age_copy.head())
# Reasignamos las filas filtradas en el DataFrame original
users_dataset[users_dataset['Age'] == 1] = under_age_copy # sets the rows according to the indexes
# users_dataset ahora refleja los cambios en las filas de menores
display(users_dataset.head())
```

```python
# Alternativa: modificar con loc en el DataFrame y eliminar NaN
# filtrar edad incorrecta (mínimo 18) Remove them from the dataset
users_dataset = pd.read_csv(ruta_users, sep='::', index_col=0,
                           # Carga con nombres de columnas y sin cabecera
                           header=None, names=['UserID','Gender','Age','Occupation','Zip-code'],
                           # Separador '::' requiere engine='python'
                           engine='python')
# Vista de 4 menores de edad antes de limpiar
display(users_dataset[users_dataset['Age'] == 1].head(4))
# loc con máscara: asigna NaN solo en filas donde Age==1
users_dataset.loc[users_dataset['Age'] == 1, 'Age'] = np.nan
# DataFrame con edades anuladas en menores
display(users_dataset)
# Filas donde Age quedó como NaN tras la asignación
display(users_dataset.loc[pd.isnull(users_dataset['Age'])].head(4))
# Detectamos valores nulos (NaN)
# drop elimina filas cuyo índice está en el conjunto de Age nulo
users_dataset.drop(users_dataset[pd.isnull(users_dataset['Age'])].index, inplace = True)
# DataFrame sin usuarios con edad inválida
display(users_dataset.head(4))
```

# 7 Ejercicios

• Hacer un análisis general de los otros dos archivos CSV en ml-1m ('movies.dat' y 'ratings.dat')
• Analizando el dataset ratings.dat, ¿hay algún usuario que no tenga ninguna review? ¿Cuántos tienen menos de 30 reviews?

```python
# Cargamos el archivo de valoraciones (ratings) del dataset MovieLens
# especificar nombres, cargar sin cabecera
import os
# Importamos Pandas: DataFrames y Series
import pandas as pd
# Ruta al archivo ratings.dat dentro de ml-1m/
ruta_ratings = os.path.join("res", "ml-1m", "ratings.dat")
# Leemos datos desde archivo
ratings_dataset = pd.read_csv(ruta_ratings, sep='::',
                              # Sin cabecera en el archivo original
                              header=None, names=['UserID','MovieID','Rating','Timestamp'],
                              # engine='python' para separador de doble dos puntos
                              engine='python')
# Tabla de valoraciones: usuario, película, nota y timestamp
display(ratings_dataset)
```

```python
# Comparamos usuarios totales vs usuarios que tienen al menos una review
# hay algún usuario que no tenga ninguna review
# groupby('UserID') agrupa por usuario; len cuenta grupos con al menos 1 rating
n_users_with_reviews = len(ratings_dataset.groupby(by='UserID'))
# Usuarios que han valorado al menos una película
print(n_users_with_reviews)
# Total de usuarios registrados en users.dat
n_users = len(users_dataset)
# Usuarios en el registro
print(n_users)
# Diferencia: usuarios sin ninguna valoración
print(n_users - n_users_with_reviews)
```

```python
# filter + lambda: usuarios con menos de 30 valoraciones
# Cuántos tienen menos de 30 reviews
# groupby agrupa por UserID; filter conserva grupos donde len(x)<30
ratings_under_30 = ratings_dataset.groupby(['UserID'])['UserID'].filter(lambda x: len(x) < 30).value_counts()
# Conteo de cuántas veces aparece cada UserID en el filtro
display(ratings_under_30)
# Número de usuarios con menos de 30 reviews
print(len(ratings_under_30))
```
