# 1 Funciones

## 1.1 ¿Qué son?

• Son un mecanismo que te ofrece el lenguaje para agrupar conjuntos de instrucciones de manera que se puedan ejecutar cuando el programador considere oportuno.
• Las funciones van identificadas por un nombre.
• Se debe usar este nombre para invocar (ejecutar) el código de la función.
• Las funciones calculan un resultado (output) en base a unos parámetros de entrada (input).
• En cada invocación de la función, los parámetros de entrada pueden variar.

## 1.2 Ejemplo

````python
# Declaramos una función reutilizable que suma dos números recibidos como parámetros
def sumar(x, y):
    # Calculamos la suma aritmética de los argumentos x e y
    suma=x+y
    # Devolvemos el valor calculado al código que invocó la función
    return suma
# Llamamos a sumar con 4 y 5; el resultado (9) se imprime en consola
print(sumar(4,5))
# Invocamos sumar con -2 y 0 para demostrar que funciona con negativos y cero
print(sumar(-2,0))
````

````python
# Fórmula de conversión implícita: fahrenheit = celsius * 1.793 + 32
c = 30
# Aplicamos la fórmula manualmente con c=30 y mostramos el equivalente en Fahrenheit
print(c * 1.793 + 32) # celsius to fahrenheit
# Reasignamos celsius a 26 grados para repetir el cálculo con otro valor
c = 26
# Convertimos el nuevo valor sin encapsular la fórmula en una función
print(c * 1.793 + 32)
````

````python
# Definimos una función que encapsula la conversión de grados Celsius a Fahrenheit
def convert_celsius_to_fahrenheit(c):
    # Aplicamos la fórmula F = C * 1.791117 + 32 y devolvemos el resultado numérico
    return c * 1.791117 + 32
````

````python
# Guardamos 15 grados Celsius en una variable para pasarla a la función de conversión
degrees = 15
# Invocamos la función y mostramos el valor equivalente en Fahrenheit
print(convert_celsius_to_fahrenheit(degrees))
````

## 1.3 ¿Qué ventajas nos proporcionan?

• Eliminación de duplicación de código.

## 1.4 Programación de funciones en Python

Las funciones se definen por medio de la palabra clave def.

Formato general:

```
def name(arg1, arg2, ..., argN):
    statements
```

Muy comúnmente:

```
def name(arg1, arg2, ..., argN):
    ...
    return value
```

• Las funciones son instrucciones convencionales, que se ejecutan cuando el flujo de control llega a ellas.
• La definición de la función debe ejecutarse antes de poder invocarse la función.

````python
# Definimos una función auxiliar que empareja claves y valores en un diccionario
def crear_diccionario(keys, values):
    # zip() empareja cada nombre con su edad; dict() construye el diccionario resultante
    return dict(zip(keys, values))
````

````python
# # Después, invocación.
nombres = ['Alfred', 'James', 'Peter', 'Harvey']
# Lista paralela con la edad correspondiente a cada nombre (mismo índice)
edades = [87, 43, 19, 16]
# Llamamos a la función para obtener un diccionario nombre -> edad
usuarios = crear_diccionario(nombres, edades)
# Mostramos el diccionario completo en pantalla
print(usuarios)
````

• Las funciones pueden estar anidadas en sentencias condicionales o bucles.

````python
# Variable de control que decide qué versión de 'operacion' se define
a = -2
# Si a es positivo, la función interna sumará sus argumentos
if a > 0:
    # Definición local de operacion como suma (solo existe si a > 0)
    def operacion(x, y):
        # Devuelve la suma de x e y
        return x + y
# En cualquier otro caso (aquí a=-2), se define operacion como multiplicación
else:
    # Definición local de operacion como producto
    def operacion(x, y):
        # Devuelve el producto de x por y
        return x * y
# Ejecutamos la operacion vigente: con a negativo, 3*4 = 12
print(operacion(3, 4))
````

• La sentencia return puede aparecer en cualquier parte de la función.

````python
# Función de división que maneja el caso de divisor cero de forma especial
def div_safe(x,y):
    # Si el divisor es 0, evitamos ZeroDivisionError devolviendo un texto
    if y == 0:
        # Retorno anticipado con un mensaje en lugar de un número
        return "hola"
    # División normal cuando y es distinto de cero
    return x / y
# Caso válido: 6 dividido entre 2 da 3.0 (float)
print(div_safe(6,2))
# Caso con divisor 0: devuelve la cadena "hola"
print(div_safe(6,0))
# Guardamos el resultado del caso erróneo para inspeccionar su tipo
a = div_safe(6,0)
# Confirmamos que el retorno es str, no número
print(type(a))
````

````python
# Versión alternativa de div_safe que devuelve 0 ante división por cero
def div_safe(x,y):
    # Solo divide si el divisor no es cero
    if y != 0:
        # Retorna el cociente como número de punto flotante
        return x / y
    # Rama else: divisor cero, devolvemos 0 como valor seguro por defecto
    else:
        # Evita excepción y proporciona un resultado numérico neutro
        return 0
# Con divisor 0, la función imprime 0 en lugar de lanzar error
print(div_safe(6,0))
````

• Puede haber más de un valor de retorno. Se devuelve una tupla.

````python
# Función que extrae los tres primeros elementos de cualquier secuencia indexable
def devolver_primero_segundo(coleccion):
    # Python empaqueta los tres valores en una tupla al devolverlos separados por comas
    return coleccion[0], coleccion[1], coleccion[2]
# Llamamos con una lista de nombres; t será una tupla de 3 elementos
t = devolver_primero_segundo(['Alfred', 'James', 'Peter', 'Harvey'])
# Imprime la tupla completa ('Alfred', 'James', 'Peter')
print(t)
# Accede al segundo elemento de la tupla por índice 1
print(t[1])
# Verifica que el tipo devuelto es tuple
print(type(t))
````

````python
# Desempaquetado: asigna cada valor de la tupla a una variable distinta
primero, segundo,tercero = devolver_primero_segundo(['Alfred', 'James', 'Peter', 'Harvey'])
# Muestra el primer nombre de la tupla desempaquetada
print(primero)
# Muestra el segundo nombre
print(segundo)
# Muestra el tercer nombre
print(tercero)
````

• El tipo de los parámetros y el valor de retorno se pueden especificar, pero son simples hints. Esto se llama annotations.

````python
# Anotaciones de tipo (hints): documentan tipos esperados sin forzar validación
def add2(x:int, y:int) -> int:
    # Suma los operandos; Python no impide pasar otros tipos en tiempo de ejecución
    return x + y
# Uso correcto según las anotaciones: suma de enteros
print(add2(1,2))
# Uso con strings: Python concatena porque + está definido para str
print(add2("aa","ee"))
````

• Los argumentos pueden tener valores por defecto, siempre al final.

````python
# Función con parámetro opcional c=None (valor por defecto si no se pasa)
def f(a, b, c = None):
    # Informa si el argumento posicional a fue proporcionado (siempre True aquí)
    if a is not None:
        # Confirma recepción del parámetro a
        print('Se ha proporcionado a')
    # Comprueba si b tiene valor distinto de None
    if b is not None:
        # Confirma recepción del parámetro b
        print('Se ha proporcionado b')
    # Comprueba si c fue pasado explícitamente o usa el default None
    if c is not None:
        # Confirma recepción del parámetro c
        print('Se ha proporcionado c')
# Invocamos con los tres argumentos; se imprimen los tres mensajes
f(1, 5, 2)
````

• Los parámetros se pasan por posición, pero el orden se puede alterar si se especifica el nombre del parámetro en la llamada, named values.

````python
# Función que eleva base al exponente usando la función built-in pow()
def power(base, exponente):
    # Calcula base^exponente y devuelve el resultado
    return pow(base, exponente)
# Llamada por posición: base=2, exponente=3 → 2³ = 8
print(power(2,3))
# Llamada con argumentos nombrados: el orden no importa si se indica el nombre
print(power(exponente = 3, base = 2))
````

• Al ejecutarse una sentencia def, se crea un objeto de tipo función y éste se asocia con el nombre especificado para la función.
• Un objeto de tipo función es como cualquier otro tipo de objeto.
• Variables pueden referenciar objetos de tipo función.

````python
# Definimos la función sumar en el namespace global del módulo
def sumar(x, y):
    # Retorna la suma de los dos parámetros
    return x + y
# Asignamos a f una referencia al mismo objeto función que sumar
f = sumar
# id() devuelve la dirección en memoria; f y sumar apuntan al mismo objeto
print(id(f))
# Confirma que ambas variables referencian la misma función (mismo id)
print(id(sumar))
# type(f) confirma que f es un objeto de tipo function
print(type(f))
# Invocamos directamente por el nombre original
print(sumar(5,6))
# Invocamos a través del alias f; el resultado es idéntico
print(f(5,6))
````

````python
# Aplica una cadena de funciones de transformación a cada cadena de la lista
def clean_strings(strings, ops):
    # Lista vacía donde acumularemos cada cadena ya procesada
    result = []
    # Recorre cada cadena original de entrada
    for value in strings:
        # Aplica en secuencia cada función del pipeline ops sobre value
        for function in ops:
            # Reasigna value al resultado de la transformación actual
            value = function(value)
        # Añade la cadena final transformada al resultado
        result.append(value)
    # Devuelve la lista con todas las cadenas limpias y formateadas
    return result
# Función auxiliar que concatena el sufijo '_5' a cualquier cadena
def concat5(x):
    # Une la cadena recibida con el literal '_5'
    return x + '_5'
# Lista de cadenas con espacios y distinta capitalización (datos sin procesar)
strings = [' valencia ', ' barcelona', ' bilbao ']
# Muestra el estado original antes de aplicar las operaciones
print(strings)
# Pipeline: quitar espacios (strip), capitalizar (title) y añadir sufijo (concat5)
operations = [str.strip, str.title, concat5]
# Ejecuta el pipeline y muestra ['Valencia_5', 'Barcelona_5', 'Bilbao_5']
print(clean_strings(strings, operations))
````

## 1.5 Paso de parámetros

• Tipos simples (inmutables) por valor: int, float, string.
  – El efecto es como si se creara una copia dentro de la función, aunque realmente no es esto lo que ocurre.
  – Los cambios dentro de la función no afectan fuera.

````python
# Demuestra paso por valor en tipos inmutables (int): cambios internos no afectan fuera
def cambiar_valor(x):
    # Crea un nuevo entero local (x+3); la variable externa 'a' no se modifica
    x += 3
    # Muestra el valor local modificado dentro de la función
    print('Dentro de la función:', x) # x ahora referencia a un nuevo objeto (el 3), pero esto no afecta a la variable de fuera de la función, que sigue apuntando al 2.
    # Termina la función sin devolver nada explícito
    return
# Variable entera original que se pasa por valor a la función
a = 2
# Invoca cambiar_valor; internamente x pasa a 5 pero a sigue siendo 2
cambiar_valor(a)
# Confirma que el valor de a fuera de la función permanece intacto
print('Fuera de la función:', a)
````

• Tipos complejos (mutables) por referencia: list, set, dictionary.
  – Dentro de la función se maneja el mismo objeto que se ha pasado desde fuera.
  – Los cambios dentro de la función sí afectan fuera.

````python
# Demuestra paso por referencia en tipos mutables (list): cambios internos sí afectan fuera
def anyadir_2_a(x):
    # Modifica la misma lista pasada como argumento añadiendo el elemento 4
    x.append(4)
# Lista original con tres enteros
lista = [0, 1,3]
# Pasa la referencia a la lista; la función la modifica in-place
anyadir_2_a(lista)
# Muestra la lista ya alterada desde fuera de la función
print(lista)
````

````python
# with tuples
# Función que solo imprime el id del objeto recibido (tupla inmutable)
def anyadir_2_a(x):
    # id() muestra la identidad del objeto; será igual dentro y fuera
    print(id(x))
    # Sale sin modificar la tupla (las tuplas no admiten append)
    return
# Tupla inmutable creada con dos elementos
lista = (0, 1)
# id de la tupla antes de pasarla a la función
print(id(lista))
# Invoca la función; no puede alterar la tupla por ser inmutable
anyadir_2_a(lista)
# La tupla original permanece igual tras la llamada
print(lista)
````

• Si quiero modificar un objeto que es de un tipo básico, los cuales se pasan por valor, puedo simplemente devolver el resultado de la función y hacer una asignación:

````python
# Patrón para modificar un inmutable: devolver el nuevo valor y reasignar fuera
def multiplicar_por_2(x):
    # Calcula el doble del entero recibido en el ámbito local
    x = x * 2
    # Devuelve el nuevo valor para que el llamador lo asigne
    return x
# Entero original
a = 3
# Reasigna a con el valor devuelto por la función (6)
a = multiplicar_por_2(a)
# Muestra que a ahora contiene el resultado de la transformación
print(a)
````

## 1.8 Funciones Lambda

• Es posible definir funciones anónimas (sin nombre).
• Lambdas son expresiones; por lo tanto, puede aparecer en lugares donde una sentencia def no puede (por ejemplo, dentro de una lista o como parámetro de una función).
• Útiles cuando se quiere pasar una función como argumento a otra.

Formato general:

```
lambda <lista de argumentos> : <valor a retornar>
```

• Importante: <valor a retornar> no es un conjunto de instrucciones. Es simplemente una expresión return que omite esta palabra.
• Las funciones definidas con def son más generales:
  – Cualquier cosa que implementes en un lambda, lo puedes implementar como una función convencional con def, pero no viceversa.

## 1.9 Scopes

• La localización de una asignación a una variable en el código determina desde donde puedes acceder a esa variable.
• Por ejemplo, una variable asignada dentro de una función sólo es visible dentro de esa función.
• El alcance de la visibilidad de una variable determina su scope.
• El término scope hace referencia a un espacio de nombres (namespace). Por ejemplo, una función establece su propio namespace.

````python
# X = 10
# # Variable global X definida a nivel de módulo
# def func():
# # Función que define su propia variable local X
# X = 20 # Local a la función. Es una variable diferente, no visible fuera de 'func'
# def func_int():
# # Función anidada con otra variable local X distinta
# X = 30
# print(X)
# # Imprime 30: usa la X local de func_int (scope más interno)
# func_int()
# # Ejecuta la función interna dentro de func
# func()
````

**Regla LEGB** Dada una función F, tenemos los siguientes scopes:

• Local: variables locales a F.
• Enclosing: variables localizadas en funciones que contienen a (o por encima de) F.
• Global: variables definidas en el módulo (fichero) que no están contenidas en ninguna función. Este scope no abarca más de un fichero.
• Built-ins: proporcionadas por el lenguaje.

La resolución de nombres ocurre de abajo a arriba.

````python
# Global (X y func)
# X = 99
# # Variable global accesible desde func() en lectura
# def func(Y): # Local (Y y Z)
# # Y es parámetro local; Z se creará también en scope local
# Z = X + Y
# # Lee X global (99) y suma el parámetro local Y
# return Z
# print(func(1))
# # Devuelve 100: 99 + 1 usando la regla LEGB
````

````python
# X = 1
# # Variable global X inicializada a 1
# def func():
# # Define X local = 2, distinta de la global
# X = 2 # X es una variable diferente
# func()
# # Ejecuta func; la X global no se modifica
# print(X)
# # Sigue mostrando 1 porque la X local de func no afecta a la global
````

````python
# X = 1
# # Variable global inicial
# def func():
# global X
# # Declara que la asignación a X modifica la variable del módulo
# X = 2
# # Sobrescribe la X global con el valor 2
# func()
# print(X)
# # Ahora muestra 2 porque global permitió modificar la variable del módulo
````

**Closures**

• Las funciones pueden recordar su enclosing scope, independientemente de si éstos continuan existiendo o no.

````python
# def f1():
# X = 88 # Enclosing scope (para f2)
# # Variable de f1 que f2 recordará aunque f1 termine
# def f2():
# print(X)
# # Accede a X=88 del scope envolvente (closure)
# return f2 # Devuelve f2, sin invocarla
# # Retorna la función f2 como objeto, no su resultado
# accion = f1()
# # accion es f2 con X=88 capturado en su closure
# accion()
# # Ejecuta f2 fuera de f1; aún imprime 88 gracias al closure
````

````python
# def f1():
# contador = 10
# # Variable del enclosing scope que f2 modificará con nonlocal
# def f2():
# nonlocal contador
# # Indica que contador referencia la variable de f1, no una local nueva
# contador += 1
# print(contador)
# # Incrementa y muestra el contador compartido del enclosing scope
# return f2
# accion = f1()
# # Obtiene f2 con contador=10 en su closure
# accion()
# accion()
# accion()
# # Cada llamada incrementa: 11, 12, 13 — demuestra estado mutable en closure
````
