# 1 Estructuras de control

## 1.1 Selección (sentencias condicionales)

• Selección de una de varias alternativas en base a alguna condición.
• Indentación para estructurar el código
• Importante el símbolo ':'

````python
# # Asigno el bvalo uno a la variable x
x = int(input("Ingrese el valor de x"))
#Implemento una condicional para saber si x es mayor a 0
if x > 0:
    # Imprime el texto o etiqueta: 'Valor positivo'
    print('Valor positivo')
# else: bloque que se ejecuta si ninguna condición previa fue verdadera
else:
    # Imprime el texto o etiqueta: 'Valor no positivo'
    print('Valor no positivo')
````

• Las condiciones son expresiones que se evalúan a True o False.
• Operadores de comparación, lógicos, de identidad y de pertenencia vistos anteriormente.

````python
# Almacena el valor entero 3 en la variable 'x'
x = 3
# Evalúa la expresión de la derecha y guarda el resultado en 'y'
y = [1, 2, 3]
# Muestra en consola el contenido de x > 0
print(x > 0)
# Muestra en consola el contenido de x > 0 and x < 10
print(x > 0 and x < 10)
# Muestra en consola el contenido de x is not y
print(x is not y)
# Muestra en consola el contenido de x in y
print(x in y)
````

• Se pueden introducir más 'ramas' en sentencias condicionales a través de la palabra clave elif.

````python
# Almacena el valor entero -3 en la variable 'x'
x = -3
# if: ejecuta el bloque indentado solo cuando x > 0
if x > 0:
    # Imprime el texto o etiqueta: 'Valor positivo'
    print('Valor positivo')
# elif: rama alternativa que se evalúa si x == 0
elif x == 0:
    # Imprime el texto o etiqueta: 'Valor nulo'
    print('Valor nulo')
# else: bloque que se ejecuta si ninguna condición previa fue verdadera
else:
    # Imprime el texto o etiqueta: 'Valor negativo'
    print('Valor negativo')
````

````python
# Almacena el valor entero 120 en la variable 'val'
val = 120
# if: ejecuta el bloque indentado solo cuando val > 0
if val > 0:
    # if: ejecuta el bloque indentado solo cuando val < 100
    if val < 100:
        # Imprime el texto o etiqueta: 'Valor positivo'
        print('Valor positivo')
    # else: bloque que se ejecuta si ninguna condición previa fue verdadera
    else:
        # Imprime el texto o etiqueta: 'Valor muy positivo'
        print('Valor muy positivo')
# else: bloque que se ejecuta si ninguna condición previa fue verdadera
else:
    # Imprime el texto o etiqueta: 'Valor negativo'
    print('Valor negativo')
````

## 2 Switch

````python
# match/case (Python 3.10+): estructura tipo switch para comparar valores
a = -6  # Valor a evaluar (negativo en este ejemplo)
match a:  # Compara el valor de 'a' contra cada case
    case _ if a > 0:  # Caso comodín con condición: a es positivo
        print('positive')  # Se ejecutaría si a > 0
    case _ if a == 0:  # Caso comodín: a es exactamente cero
        print('zero')  # Se ejecutaría si a == 0
    case _ if a < 0:  # Caso comodín: a es negativo
        print('negative')  # Se ejecuta porque a = -6
````

````python
# match/case para mapear letras de días de la semana a su nombre en español
a = 'l'  # Letra del día (minúscula; los case usan mayúsculas)
match a:  # Evalúa el valor de 'a' contra cada patrón
    case 'L':  # Si la letra es 'L' (Lunes)
        print('Lunes')  # Imprime el nombre del lunes
    case 'M':  # Si la letra es 'M' (Martes)
        print('Martes')  # Imprime el nombre del martes
    case 'X':  # Si la letra es 'X' (Miércoles)
        print('Miercoles')  # Imprime el nombre del miércoles
    case _:  # Caso por defecto: ningún patrón anterior coincidió
        print('Wrong Day')  # Se ejecuta porque 'l' ≠ 'L', 'M' ni 'X'
````

## 3 Iteración (bucles)

• Repetición de un bloque de código.
• La terminación del bucle depende del tipo de bucle.
• Dos tipos: while y for.

### Bucle 'while'

• Repetición de un bloque de código hasta que se deje cumplir una expresión (es decir, hasta que una condición evalue a False).
• Si la condición evalua a False desde el principio, el bloque de código nunca se ejecuta.
• Cuidado con los bucles infinitos.

Formato general:

```
while test: # Mientras se cumple la condición
    statements # Instrucciones a ejecutar
```

````python
# # Ejemplo: Mostrar los primeros 3 objetos de una lista
indice = 0
# Evalúa la expresión de la derecha y guarda el resultado en 'numeros'
numeros = [9, 4, 7, 1, 2]
# Bucle while: repite el bloque mientras indice < 3
while indice < 3:
    # Muestra en consola el contenido de numeros[indice]
    print(numeros[indice])
    # Actualiza 'indice' sumando un valor (acumulador)
    indice =indice+1
````

````python
# Ejemplo: recorrer una lista con while, filtrar valores < 10 y contarlos
numeros = [33, 3, 9, 21, 1, 7, 12, 10, 8]  # Lista de números a analizar
contador = 0  # Acumulador de cuántos números son menores que 10
indice = 0  # Posición actual dentro de la lista
while indice < len(numeros):  # Mientras queden elementos por revisar
    if numeros[indice] < 10:  # Comprueba si el número en esa posición es < 10
        print(numeros[indice])  # Muestra el número que cumple la condición
        contador += 1  # Incrementa el contador de números menores que 10
    indice += 1  # Avanza al siguiente índice (siempre, haya o no cumplido el if)
print('Contador:', contador)  # Total de números menores que 10 encontrados
print('Indice:', indice)  # Índice final (= longitud de la lista)
````

````python
# Guarda la cadena de texto 'Pablo' en 'nombre'
nombre = 'Pablo'
# Bucle while: repite el bloque mientras nombre
while nombre: # Mientras 'nombre' no sea vacío
    # Muestra en consola el contenido de nombre
    print(nombre)
    # Extrae un subconjunto por slicing y lo guarda en 'nombre'
    nombre = nombre[1:]
````

````python
# Bucle while controlado: imprime del 0 al 9 evitando bucle infinito
i = 0  # Contador inicial del bucle
while i < 10:  # Repite mientras i sea menor que 10
    print(i)  # Muestra el valor actual de i en cada iteración
    i += 1  # Incrementa i en 1 para avanzar y eventualmente salir del bucle
````

### Bucle 'for'

• Permite recorrer los items de una sequencia o un objeto iterable.
• Funciona en strings, listas, tuplas, etc.

Formato general:

```
for item in objeto: # Asigna los items del objeto a la variable item en cada iteración
    statements # Instrucciones a ejecutar
```

````python
# Evalúa la expresión de la derecha y guarda el resultado en 'peliculas'
peliculas = ['Matrix', 'The purge', 'Avatar', 'Star Wars']
# Bucle for: itera con 'pelicula in peliculas' en cada vuelta
for pelicula in peliculas:
    # Muestra en consola el contenido de pelicula
    print(pelicula)
````

• También se usa para iterar un número preestablecido de veces (counted loops):

````python
# Bucle for: itera con 'i in range(10)' en cada vuelta
for i in range(10):
    # Muestra en consola el contenido de i
    print(i)
````

• range es útil en combinación con len porque permite acceder a los elementos de una secuencia por posición.

````python
# Guarda la cadena de texto 'Pablo' en 'nombre'
nombre = 'Pablo'
# Bucle for: itera con 'i in range(len(nombre))' en cada vuelta
for i in range(len(nombre)):
    # Muestra en consola el contenido de nombre[i]
    print(nombre[i])
````

````python
# Bucle for: itera con 'letra in nombre' en cada vuelta
for letra in nombre:
    # Muestra en consola el contenido de letra
    print(letra)
````

• Iteración de tuplas:

````python
# Evalúa la expresión de la derecha y guarda el resultado en 'tuplas'
tuplas = [(1, 2,4), (3, 4), (5, 6)]
# Bucle for: itera con 'a, b,c in tuplas' en cada vuelta
for a, b,c in tuplas:
    # Muestra en consola el contenido de a, b,c
    print(a, b,c)
````

• Iteración de diccionarios. Se iteran las claves:

````python
# Creamos o usamos un diccionario
diccionario = {'a': 1, 'b': 2, 'c': 3}
# Bucle for: itera con 'key in diccionario' en cada vuelta
for key in diccionario:
    # Muestra mensaje formateado en consola: f"{key} => {diccionario[key]}"
    print(f"{key} => {diccionario[key]}")
````

• Para iterar los pares (clave-valor) o únicamente los valores, se deben usar los métodos items y values.

````python
# Bucle for: itera con 'key, value in diccionario.items()' en cada vuelta
for key, value in diccionario.items():
    # Muestra mensaje formateado en consola: f"{key} => {value}"
    print(f"{key} => {value}")
````

````python
# Bucle for: itera con 'value in diccionario.values()' en cada vuelta
for value in diccionario.values():
    # Muestra en consola el contenido de value
    print(value)
````

• La variable item en la cabecera del for puede ser cualquier expresión que sea válida como parte izquierda de una asignación convencional.

````python
# Bucle for: itera con 'a, b, c in [(1, 2, 3), (4, 5, 6)]' en cada vuelta
for a, b, c in [(1, 2, 3), (4, 5, 6)]:
    # Muestra en consola el contenido de a, b, c
    print(a, b, c)
````

## 3.1 Las sentencias break, continue y else

### Break y continue

• Solo tienen sentido dentro de bucles.
• Break permite terminar el bucle por completo.
• Continue permite saltar a la siguiente iteración, continuando con el bucle.
• Pueden aparecer en cualquier parte de un bucle, pero normalmente aparecen dentro de sentencias condicionales (if).

````python
# # Ejemplo: break
rating_to_find = 4.2
# Evalúa la expresión de la derecha y guarda el resultado en 'movie_ratings'
movie_ratings = [4.3, 2.5, 1.7, 4.2, 3.8, 3.3, 4.5]
# Bucle for: itera con 'rating in movie_ratings' en cada vuelta
for rating in movie_ratings:
    # Muestra en consola el contenido de rating
    print(rating)
    # if: ejecuta el bloque indentado solo cuando rating == rating_to_find
    if rating == rating_to_find:
        # Imprime el texto o etiqueta: "Found"
        print("Found")
        # break: detiene el bucle en cuanto se encuentra el resultado buscado
        break
    # Muestra en consola el contenido de rating
    print(rating)
````

````python
# Bucle for: itera con 'i in range(10)' en cada vuelta
for i in range(10):
    # if: ejecuta el bloque indentado solo cuando i % 2 == 0
    if i % 2 == 0:
        # Saltamos a la siguiente iteración del bucle
        continue # Si el número es par, salta a la siguiente iteración
    # Muestra mensaje formateado en consola: f'Odd number {i}'
    print(f'Odd number {i}')
````

• Observa como la sentencia continue te puede ayudar a reducir el número de niveles de anidamiento.
• Sin continue el anterior ejemplo sería:

````python
# Bucle for: itera con 'i in range(10)' en cada vuelta
for i in range(10):
    # if: ejecuta el bloque indentado solo cuando i % 2 != 0
    if i % 2 != 0:
        # Imprime el texto o etiqueta: 'Numero impar:', end=' '
        print('Numero impar:', end=' ')
        # Muestra en consola el contenido de i
        print(i)
````

````python
# Bucle for: itera con 'i in range(5)' en cada vuelta
for i in range(5):
    # Bucle for: itera con 'a in range(2)' en cada vuelta
    for a in range(2):
        # Muestra mensaje formateado en consola: f"{i} - {a}"
        print(f"{i} - {a}")
        # if: ejecuta el bloque indentado solo cuando i == 3 and a == 0
        if i == 3 and a == 0:
            # break: detiene el bucle en cuanto se encuentra el resultado buscado
            break
````

### Else

• Los bucles pueden tener una sensencia else.
• Resulta poco intuitiva para muchos programadores porque esta sintaxis no existe en otros lenguajes.
• Se ejecuta cuando el bucle termina con normalidad; es decir, cuando no termina a causa de un break.

````python
# Evalúa la expresión de la derecha y guarda el resultado en 'lista'
lista = [60,60,30,60]
# Almacena el valor entero 60 en la variable 'element_to_find'
element_to_find = 60
# Bucle for: itera con 'element in lista' en cada vuelta
for element in lista:
    # if: ejecuta el bloque indentado solo cuando element == element_to_find
    if element == element_to_find:
        # Imprime el texto o etiqueta: "found"
        print("found")
        # break: detiene el bucle en cuanto se encuentra el resultado buscado
        break
# else: bloque que se ejecuta si ninguna condición previa fue verdadera
else:
    # Imprime el texto o etiqueta: "not found"
    print("not found")
````

````python
# with flag
lista = [10,30,50,40]
# Almacena el valor entero 30 en la variable 'element_to_find'
element_to_find = 30
# Bandera 'found' en False: indica que aún no se ha encontrado el objetivo
found = False
# Bucle for: itera con 'element in lista' en cada vuelta
for element in lista:
    # if: ejecuta el bloque indentado solo cuando element == element_to_find
    if element == element_to_find:
        # Activa la bandera 'found' a True cuando se cumple la condición buscada
        found = True
        # break: detiene el bucle en cuanto se encuentra el resultado buscado
        break
# if: ejecuta el bloque indentado solo cuando found
if found:
    # Imprime el texto o etiqueta: "dato encontrado"
    print("dato encontrado")
# else: bloque que se ejecuta si ninguna condición previa fue verdadera
else:
    # Imprime el texto o etiqueta: "not found"
    print("not found")
````

# 4 List Comprehensions

• Permiten construir listas a través de la ejecución repetida (sentencia for) de una expressión para cada item de un objeto iterable.
• Van entre '[' y ']'. Esto es indicativo de que estamos construyendo una lista.

Sintaxis:

```
[<expression> for <item> in <iterable>]
```

Ejemplo: repetir los carácteres de un string.

````python
# List comprehension: triplica cada carácter del nombre 'Enrique'
[char * 3 for char in 'Enrique']  # char*3 repite cada letra 3 veces: 'EEE', 'nnn', ...
````

````python
# Equivalente con bucle for: construir lista de caracteres uno a uno
lista = []  # Lista vacía que iremos rellenando
for char in 'Enrique':  # Recorre cada letra del string
    lista.append(char)  # Añade la letra al final de la lista
print(lista)  # Muestra ['E', 'n', 'r', 'i', 'q', 'u', 'e']
````

Comúnmente, el item de la sentencia for aparecerá en la expresión principal, pero eso no es obligatorio.

````python
# List comprehension: genera un 2 por cada carácter (el item no se usa en la expresión)
[2 for _ in 'Enrique']  # '_' indica que ignoramos el carácter; hay 7 letras → 7 doses
````

### 4.0.1 Versión extendida

Se puede especificar un filtro (sentencia if ) para obtener únicamente los elementos que cumplan cierta condición.

Sintaxis:

```
[<expression> for <item> in <iterable> if <condition>]
```

Ejemplo: obtener números pares:

````python
# Comprehension: construye la colección aplicando una expresión a cada elemento iterable
[y for y in range(9) if y % 2 == 0]
````

• Las list comprehension no son realmente requeridas, ya que siempre podemos escribir un bucle equivalente.

````python
# Bucle equivalente a la list comprehension de números pares del 0 al 8
pares = []  # Lista vacía para almacenar los números pares
for x in range(9):  # Recorre los enteros del 0 al 8
    if x % 2 == 0:  # Comprueba si x es divisible entre 2 (par)
        pares.append(x)  # Añade el número par a la lista
print(pares)  # Muestra [0, 2, 4, 6, 8]
````

### 4.0.2 Versión completa

La sentencia if de una list comprehension también puede contener una expresión alternativa.

Sintaxis:

```
[<expression_1> if <condition> else <expression_2> for <item> in <iterable>]
```

Ejemplo: poner a cero los números pares.

````python
# Comprehension: construye la colección aplicando una expresión a cada elemento iterable
[0 if x % 2 == 0 else x for x in range(9)]
````

````python
# Comprehension: construye la colección aplicando una expresión a cada elemento iterable
[x**2 if x > 2 else x for x in range(-4,5) if x > 0]
````

````python
# Bucle con if anidado: eleva al cuadrado solo los positivos mayores que 2
lista = []  # Lista resultado
for x in range(-4, 5):  # Recorre enteros de -4 a 4 inclusive
    if x > 0:  # Solo procesa números estrictamente positivos
        if x > 2:  # Si además es mayor que 2, guardamos su cuadrado
            lista.append(x ** 2)  # Añade x² (9, 16 para x=3,4)
        else:  # Si es positivo pero ≤ 2, guardamos x tal cual
            lista.append(x)  # Añade 1 o 2 sin elevar
print(lista)  # Muestra [1, 2, 9, 16]
````

````python
# Comprehension: construye la colección aplicando una expresión a cada elemento iterable
[x**2 if x > 0 else 100 if x==0 else x for x in range(-4,5)]
````

````python
# Bucle con if/elif/else: distinto tratamiento según signo de x
lista = []  # Lista resultado
for x in range(-4, 5):  # Recorre enteros de -4 a 4
    if x > 0:  # Números positivos → se guarda su cuadrado
        lista.append(x ** 2)
    elif x == 0:  # El cero se sustituye por 100
        lista.append(100)
    else:  # Números negativos → se guardan sin modificar
        lista.append(x)
print(lista)  # Muestra [-4, -3, -2, -1, 100, 1, 4, 9, 16]
````

### 4.0.3 Anidamiento

Las list comprehensions soportan anidamiento en sus expresiones.

Ejemplo: bucle anidado para obtener una lista de tuplas que combinan los elementos de dos listas dadas.

````python
# List comprehension anidada: producto cartesiano de dos listas como tuplas
lista_1 = [1, 2, 3]  # Primera lista de valores
lista_2 = [4, 5]  # Segunda lista de valores
[(x, y) for x in lista_1 for y in lista_2]  # Combina cada x con cada y: 3×2=6 tuplas
````

### 4.0.4 Pros y contras

• Principales ventajas de las comprehensions en comparación a un bucle convencional:
  – Expresión compacta y legible, si estás familiarizado con la sintaxis.
  – Mejor rendimiento.
• Desventaja: no escalan bien. Una list comprehension se puede convertir rápidamente en una expresión difícil de entender.

### 4.0.5 Otros tipos de comprehensions

**Dictionary comprehensions**

Usando '{' y '}' como delimitadores y una expresión 'clave : valor', se obtiene un diccionario en lugar de una lista.

Ejemplo: diccionario donde cada valor es el cuadrado de la clave.

````python
# Asigna a 'd' la secuencia generada por range(10)}
d = {x : x*x for x in range(10)}
# Muestra en consola el contenido de d
print(d)
# Indica el tipo Python del objeto d
print(type(d))
````

````python
# Ejercicio 1: sumar todos los elementos de una lista numérica
numeros = [1, 5, 9, 2, 3]  # Lista de números a sumar
suma = 0  # Acumulador inicializado en cero
for numero in numeros:  # Recorre cada elemento de la lista
    suma += numero  # Suma el elemento actual al acumulador
print(suma)  # Muestra el total: 20
````

````python
# Ejercicio 2: eliminar duplicados preservando el orden de aparición
numeros = [1, 2, 2, 3, 4, 5, 4, 5, 3, 3, 1]  # Lista con elementos repetidos
resultado = []  # Lista vacía para guardar valores únicos
for numero in numeros:  # Recorre cada número de la lista original
    if numero not in resultado:  # Solo añade si aún no está en resultado
        resultado.append(numero)  # Añade el número por primera vez
print(resultado)  # Muestra [1, 2, 3, 4, 5] sin duplicados
````

````python
# Ejercicio 3: diccionario {n: n²} para n desde 1 hasta n dado
n = 5  # Límite superior del rango
diccionario = {}  # Diccionario vacío que iremos rellenando
for i in range(1, n + 1):  # Recorre los enteros del 1 al 5 inclusive
    diccionario[i] = i * i  # Asocia cada clave i con su cuadrado i²
print(diccionario)  # Muestra {1: 1, 2: 4, 3: 9, 4: 16, 5: 25}
````

````python
# Ejercicio 4
palabras = ["", "Valencia", "python", "asignaturas", "programación", "java", "estudiantes", "académicos"]
# Bucle for: itera con 'palabra in palabras' en cada vuelta
for palabra in palabras:
    # if: ejecuta el bloque indentado solo cuando len(palabra) >= 9 and palabra[0] == 'a'
    if len(palabra) >= 9 and palabra[0] == 'a':
        # Muestra mensaje formateado en consola: f"Palabra encontrada: {palabra}"
        print(f"Palabra encontrada: {palabra}")
        # break: detiene el bucle en cuanto se encuentra el resultado buscado
        break
# else: bloque que se ejecuta si ninguna condición previa fue verdadera
else:
    # Imprime el texto o etiqueta: "Palabra no encontrada"
    print("Palabra no encontrada")
````

````python
# Ejercicio 5: índices donde L[i] es múltiplo de 3
L = [3, 5, 13, 12, 1, 9]  # Lista de números positivos
resultado = []  # Lista para guardar los índices que cumplen la condición
for i in range(len(L)):  # Recorre cada posición i de la lista
    if L[i] % 3 == 0:  # Comprueba si el valor en posición i es múltiplo de 3
        resultado.append(i)  # Guarda el índice i en la lista resultado
print(resultado)  # Muestra [0, 3, 5] porque L[0]=3, L[3]=12, L[5]=9
````

````python
# Ejercicio 6
d = {'a':4.3, 'b':1, 'c':7.8, 'd':-5}
# Evalúa la expresión de la derecha y guarda el resultado en 'maximo'
maximo = float("-inf") # Inicializamos la variable 'maximo' a -infinito. Para este ejercicio, también valdría cualquier valor negativo muy pequeño.
# Obtenemos la longitud/cantidad de elementos
if(len(d) == 0):
    # Imprime el texto o etiqueta: "El diccionario está vacío"
    print("El diccionario está vacío")
# Bucle for: itera con 'clave, valor in d.items()' en cada vuelta
for clave, valor in d.items():
    # if: ejecuta el bloque indentado solo cuando valor > maximo
    if valor > maximo:
        # Evalúa la expresión de la derecha y guarda el resultado en 'maximo'
        maximo = valor
        # Evalúa la expresión de la derecha y guarda el resultado en 'resultado'
        resultado = clave
# Muestra en consola el contenido de resultado
print(resultado)
````

````python
# Ejercicio 7
a = [2, 4, 6, 8]
# Evalúa la expresión de la derecha y guarda el resultado en 'b'
b = [7, 11, 15, 22]
# Bucle for: itera con 'i in a' en cada vuelta
for i in a:
    # if: ejecuta el bloque indentado solo cuando i > 5
    if i > 5:
        # Bucle for: itera con 'j in b' en cada vuelta
        for j in b:
            # if: ejecuta el bloque indentado solo cuando j < 14
            if j < 14:
                # Muestra mensaje formateado en consola: f"{i} x {j} = {i*j}"
                print(f"{i} x {j} = {i*j}")
````
